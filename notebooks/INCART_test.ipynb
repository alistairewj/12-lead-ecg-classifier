```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
import re
import numpy as np
import pickle
from IPython.display import Image
from collections import OrderedDict

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
root_dir = '../experiments'
all_dirs = os.listdir(root_dir)
sample_dirs = sorted([d for d in all_dirs if 'sample' in d])
my_order = [sample_dirs[3], sample_dirs[1], sample_dirs[0], sample_dirs[2]]
data_dict = {'split_4_1': dict.fromkeys(my_order, {}), 
             'split_4_2': dict.fromkeys(my_order, {}), 
             'split_4_3': dict.fromkeys(my_order, {}), 
             'split_4_4': dict.fromkeys(my_order, {})}

for k,v in data_dict.items():
    print(k,v)
print()

for key, value in data_dict.items():
    for k, v in value.items():
        path = os.path.join(root_dir, k, key)
        files = os.listdir(path)
        history_file = [file for file in files if 'history' in file][0]
        roc_file = [file for file in files if 'roc' in file][0]
        new_dict = {
            'history': history_file,
            'rock_file': roc_file
        }
        data_dict[key][k].update(new_dict)
    
for k,v in data_dict.items():
    print(k,v)
print()
```

```python
# Files from training phase
train_files = sorted([file for file in os.listdir(os.path.join(root_dir, 'train_stratified')) if 'split_4' in file])
ROC_dirs = [os.path.join(root_dir, 'train_stratified', file) for file in train_files if 'ROC' in file]
ROC_files = [os.path.join(file, 'roc-e50.png') for file in ROC_dirs if 'ROC' in file]
train_history_files = [os.path.join(root_dir, 'train_stratified', file) for file in train_files if 'history' in file]
```

```python
transformers = ['ValClip()', 'RandomClip()']*2

for j, (d, values) in enumerate(data_dict.items()):
    print(color.GREEN + color.BOLD + '='*30, d.upper(), '='*30 + color.END, '\n')
    print(color.RED + color.BOLD + '='*25, 'VALIDATION PHASE', '='*25 + color.END, '\n')
    print(color.RED +'{:>30}: {:>5}'.format('Transform', 'ValClip()') + color.END)
    # train stuff -----------------
    with open(train_history_files[j], 'rb') as history_file:
                    while True:
                        try:
                            train_history = pickle.load(history_file)
                        except EOFError:
                            break
    
    for key, value in sorted(train_history.items()):
        if 'csv' in key:
            print(color.BOLD + color.GREEN + '{:>30}: {:>5}'.format(key, os.path.basename(value)) + color.END)
        elif 'labels' in key:
            continue
        elif 'val' in key and not 'loss' in key:
            print('{:>30}: {:>5}'.format(' '.join(key.split('_')), round(value[-1], 4)))

    display(Image(ROC_files[j], width=700))
    
    # test stuff ------------------
    print(color.RED + color.BOLD + '='*25, 'TESTING PHASE', '='*25 + color.END, '\n')
    for i, (k, v) in enumerate(values.items()):
        split_d = os.path.join(root_dir, k, d)
        print(color.RED +'{:>30}: {:>5}'.format('Directory', k) + color.END)
        print(color.RED +'{:>30}: {:>5}'.format('Transform', transformers[i]) + color.END)

        for var, file in v.items():
            true_path = os.path.join(split_d, file)
            if 'history' in var:
                with open(true_path, 'rb') as history_file:
                    while True:
                        try:
                            test_history = pickle.load(history_file)
                        except EOFError:
                            break

                for key, value in sorted(test_history.items()):
                    if 'csv' in key:
                        print(color.BOLD + color.GREEN + '{:>30}: {:>5}'.format(key, os.path.basename(value)) + color.END)
                    elif 'labels' in key:
                        continue
                    else:
                        print('{:>30}: {:>5}'.format(' '.join(key.split('_')[1:4]), round(value, 4)))
                print()
                
            if 'roc' in true_path:
                display(Image(true_path, width=700))
        print('--'*40)
      
```
